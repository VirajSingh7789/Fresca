```python
from colorthief import ColorThief
import os
import glob
import numpy as np
import pandas as pd
```

```python
# p a n t

colors = [] * 0
names = [] * 0
for filepath in glob.iglob(r'C:\Users\edwardchen\Data Science\HackGT2020\Uniqlo\Pant\*.jpg'):
    count = 0
    color_thief = ColorThief(filepath)
    dominant_color = color_thief.get_color(quality=1)
    palette = color_thief.get_palette(color_count=2)
    for color in palette[0]:
        if color > 210 and color < 230:
            count+=1
        if count == 3:
            palette[0] = palette[1]
    name = filepath[56:].replace(".JPG", "")
    colors.append(palette[0])
    names.append(name)
    
red = [] * 0
green = [] * 0
blue = [] * 0

for color in colors:
    red.append(color[0])
    green.append(color[1])
    blue.append(color[2])
    
prices = [] * 0
for filepath in glob.iglob(r'C:\Users\edwardchen\Data Science\HackGT2020\Uniqlo\Pant\*.jpg'):
    filepath = filepath.replace("-", " ")
    filepath = filepath.replace(".", " ")
    res = [int(i) for i in str(filepath).split() if i.isdigit()]
    price = (res[0]) + (res[1]) * 0.01
    prices.append(price)

pant_type = [] * 0
for p in range(len(red)):
    pant_type.append('P a n t')
    
name_data = np.array(names)
price_data = np.array(prices)
red_data = np.array(red)
green_data = np.array(green)
blue_data = np.array(blue)
pant_data = np.array(pant_type)

name_series = pd.Series(name_data)
price_series = pd.Series(price_data)
red_series = pd.Series(red_data)
green_series = pd.Series(green_data)
blue_series = pd.Series(blue_data)
pant_series = pd.Series(pant_data)

pant_dataframe = pd.concat([name_series, price_series, red_series, green_series, blue_series, pant_series], axis = 1)
pant_dataframe_final = pant_dataframe.rename({0: 'Name', 1: 'Price', 2: 'Red', 3: 'Green', 4: 'Blue', 5: 'Type'}, axis=1)
pant_dataframe_final
```
